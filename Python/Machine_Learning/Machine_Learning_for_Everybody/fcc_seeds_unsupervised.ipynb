```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cols = ["area", "perimeter", "compactness", "length", "widht", "asymmetry", "groove", "class"]
df = pd.read_csv("/content/drive/MyDrive/data/seeds_dataset.txt", names = cols, sep = "\s+")
```

```python
df.head()
```

```python
for i in range(len(cols) - 1):
    for j in range(i + 1, len(cols) - 1):
        x_label = cols[i]
        y_label = cols[j]
        sns.scatterplot(
            x = x_label,
            y = y_label,
            data = df,
            hue = "class"
        )
        plt.show()
```

# Clustering

```python
from sklearn.cluster import KMeans
```

```python
x = "compactness"
y = "asymmetry"
X = df[[x, y]].values
```

```python
kmeans = KMeans(n_clusters=3).fit(X)
```

```python
clusters = kmeans.labels_
clusters
```

```python
df["class"].values
```

```python
cluster_df = pd.DataFrame(np.hstack((X, kmeans.labels_.reshape(-1, 1))), columns=[x, y, "class"])
```

## K Means classes

```python
sns.scatterplot(x=x, y=y, hue="class", data=cluster_df)
plt.plot()
```

## Original classes

```python
sns.scatterplot(x=x, y=y, hue="class", data=df)
plt.plot()
```

# Higher Dimensions

```python
X = df[cols[:-1]].values
```

```python
kmeans = KMeans(n_clusters = 3).fit(X)
cluster_df = pd.DataFrame(np.hstack((X, clusters.reshape(-1, 1))), columns = df.columns)
```

```python
# K Means classes
sns.scatterplot(x=x, y=y, hue="class", data=cluster_df)
plt.plot()
```

```python
# Original classes
sns.scatterplot(x=x, y=y, hue="class", data=df)
plt.plot()
```

# PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
transformed_x = pca.fit_transform(X)
```

```python
X.shape
```

```python
transformed_x.shape
```

```python
transformed_x[:5]
```

```python
plt.scatter(transformed_x[:,0], transformed_x[:,1])
plt.show()
```

```python
kmeans_pca_df = pd.DataFrame(np.hstack((transformed_x, kmeans.labels_.reshape(-1, 1))), columns=["pca1", "pca2", "class"])
```

```python
truth_pca_df = pd.DataFrame(np.hstack((transformed_x, df["class"].values.reshape(-1, 1))), columns=["pca1", "pca2", "class"])
```

```python
# K Means classes
sns.scatterplot(x="pca1", y="pca2", hue="class", data=kmeans_pca_df)
plt.plot()
```

```python
# Thruth classes
sns.scatterplot(x="pca1", y="pca2", hue="class", data=truth_pca_df)
plt.plot()
```

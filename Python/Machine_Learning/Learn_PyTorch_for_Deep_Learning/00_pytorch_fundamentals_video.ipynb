### Matrix multiplication

Two main ways of performing multiplication in neural networks and deep learning:

1. Element-wise multiplication
2. Matrix multiplication (dot product)

[More information on multiplying matrices](https://www.mathsisfun.com/algebra/matrix-multiplying.html)

[matrixmultiplication.xyz](http://matrixmultiplication.xyz)

There are two main rules that performing matrix multiplication needs to satisfy:

1. The **inner dimensions** must match:
* `(3, 2 ) @ (3, 2)` won't work
* `(2, 3) @ (3, 2)` will work
* `(3, 2) @ (2, 3)` will work
2. The resulting matrix has the shape of the **outer dimensions**:
* `(2, 3) @ (3, 2)` -> `(2, 2)`
* `(3, 2) @ (2, 3)` -> `(3, 3)`

```python
#torch.matmul(torch.rand(3, 2), torch.rand(3, 2))
```

```python
torch.matmul(torch.rand(3, 2), torch.rand(2, 3))
```

### One of the most common errors in deep learning: shape errors

```python
# Shapes for matrix multiplication
tensor_A = torch.tensor([
    [1, 2],
    [3, 4],
    [5, 6]
])

tensor_B = torch.tensor([
    [7, 10],
    [8, 11],
    [9, 12]
])

#torch.mm(tensor_A, tensor_B) # torch.mm is the same as torch.matmul
```

```python
tensor_A.shape, tensor_B.shape
```

To fix our tensor shape issues, we can manipulate the shape of one of our tensors using a **transpose**.

A **transpose** switches the axes or dimensions of a given tensor

```python
tensor_B
```

```python
tensor_B.T, tensor_B.T.shape
```

```python
# The matrix multiplication operation works when tensor_B is transposed
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape }")
print(f"New shapes: tensor_A = {tensor_A.shape} (same shape as above), tensor_B.T = {tensor_B.T.shape}")
print(f"Multiplying: {tensor_A.shape} @ {tensor_B.T.shape} <- inner dimensions must match")
print("Output:\n")
output = torch.matmul(tensor_A, tensor_B.T)
print(output)
print(f"\nOutput shape: {output.shape}")
```

## Reshaping, stacking, squeezing and usqueezing tensors

* Reshaping - reshapes an input tensor to a defined shape
* View - Return a view of an input tensor of certain shape but keep the same memory as the origina tensor
* Stacking - combine multiple tensors on top of each other (vstack) or side by side (hstack)
* Sqeeze - remove all `1` dimensions from a tensor
* Unsqueeze - add a `1` dimension to a target tensor
* Permute - Return a view of the input with dimensions permuted (swapped) in a certain way

```python
x = torch.arange(1., 11.)
x, x.shape
```

```python
# Add an extra dimension
#x_reshaped = x.reshape(1, 7)
```

```python
# The original tensor has 10 element!!!
x_reshaped = x.reshape(1, 10)
x_reshaped, x_reshaped.shape
```

```python
x_reshaped = x.reshape(10, 1)
x_reshaped, x_reshaped.shape
```

```python
x_reshaped = x.reshape(5, 2)
x_reshaped, x_reshaped.shape
```

```python
x_reshaped = x.reshape(1, 10)
x_reshaped, x_reshaped.shape
```

```python
# Change the view
z = x.view(1, 10)
z, z.shape
```

```python
x
```

```python
# Changing z changes x (because a view of a tensor shares the same memory as the original input)
z[:, 0] = 5
z, x # x also changed!!!
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=0)
x_stacked
```

```python
x_stacked_v = torch.stack([x, x, x, x], dim=1)
x_stacked_v
```

```python
# torch.squeeze() - removes all single dimensions from a target tensor
print(f"Previous tensor: {x_reshaped}")
print(f"Previous shape: {x_reshaped.shape}")

# Remove extra dimensions from x_reshape
x_sqeezed = x_reshaped.squeeze()
print(f"\nNew tensor: {x_sqeezed}")
print(f"New shape: {x_sqeezed.shape}")
```

```python
# torch.unsqueeze() - adds a single dimension to a target tensor at a specific dim (dimension)
print(f"Previous target: {x_sqeezed}")
print(f"Previous shape: {x_sqeezed.shape}")

# Add an extra dimension with unsqueeze()
x_unsqueezed = x_sqeezed.unsqueeze(dim=0)
print(f"\nNew tensor: {x_unsqueezed}")
print(f"New shape: {x_unsqueezed.shape}")
```

```python
x_unsqueezed_v = x_sqeezed.unsqueeze(dim=1)
print(f"\nNew tensor: {x_unsqueezed_v}")
print(f"New shape: {x_unsqueezed_v.shape}")
```

```python
# torch.permute() - rearranges the dimensions of a target tensor in a specified order
x_original = torch.rand(size=(224, 224, 3)) # [height, width, colour_channels]
x_original, x_original.shape
```

```python
# Permute the original tensor to rearrange the axis (or dim) order
x_permuted = x_original.permute(2, 0, 1) # [colour channel, height, width]
x_permuted, x_permuted.shape
```

```python
x_original[0, 0, 0] = -5
x_original[0, 0, 0], x_permuted[0, 0, 0] # permuted returns a view
```

## Reproducibility (trying to take random out of random)

In short  how a neural netrwork learns:  
`start with a random numbers -> tensor operations -> update random numbers to try and make them better representations of the data -> again -> again -> again...`

To reduce the randomness in neaural networks and PyTorch comes the concept of a **random seed**.

Essentially what the random seed does is "flavour" the randomness.

```python
random_tensor_A = torch.rand(3, 4)
random_tensor_B = torch.rand(3, 4)

print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
```

```python
# Set the random seed
RANDOM_SEED = 42

torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3, 4)

torch.manual_seed(RANDOM_SEED)
random_tensor_D = torch.rand(3, 4)

print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
```
